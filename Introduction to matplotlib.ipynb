```python
heart_disease = pd.read_csv("heart_disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot(kind="hist");
```

```python
heart_disease.plot(kind="hist");
```

```python
over_50  = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
fig, ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
```

```python
#OO maethod from scratch

fig, ax=plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);

#Detailing over graph
ax.set(title="CHOLESTROL LEVEL WIHT AGE GROUP", xlabel="age group", ylabel="cholestrol");

#adding a legend
ax.legend(*scatter.legend_elements(), title="legends");

#adding a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
plt.style.use('seaborn-deep')

#adding 2 axis in the graph
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True);
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap='winter');
ax0.set(title="CHOLESTROL LEVEL", ylabel="cholestrol level");
ax0.set_xlim([50, 80])
ax0.legend(*scatter.legend_elements(), title="legend");
ax0.axhline(over_50["chol"].mean(), linestyle="--");

scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap='bwr');
ax1.set(title="MAX HEART RATE WITH AGE", xlabel="AGE OF PERSONS", ylabel="MAX HEART RATE");
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
ax1.legend(*scatter.legend_elements(), title="legend");
ax1.axhline(over_50["thalach"].mean(), linestyle="--");
fig.suptitle("HEART DISEASE ANALYSIS", fontsize=20, fontweight="bold");
```

## Manipulating data

```python
car_sales["Make"] = car_sales["Make"].str.lower()
car_sales
```

```python
missing_data = pd.read_csv("missing_data.csv")
missing_data
```

```python
missing_data["Odometer"].fillna("No data availabel", inplace = True)
missing_data
```

```python
missing_data["Price"].fillna("Not listed", inplace = True)
missing_data
```

```python
missing_data_dropped = missing_data.dropna(inplace = True)
missing_data
```

```python
seats_column = pd.Series([5,5,5,5,5,5])
car_sales["Seats"] = seats_column
car_sales["Seats"].fillna(5, inplace = True)
car_sales
```

```python
fuel_economy_column = pd.Series([12,12,14,45,12])
car_sales["Fuel_Economy/Litre"] = fuel_economy_column
car_sales["Fuel_Economy/Litre"].fillna(12, inplace = True)
car_sales
```

```python
car_sales["Total_Fuel_Used(L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel_Economy/Litre"]
car_sales
```

```python
car_sales["Passed road safety"] = True
car_sales
```

```python
car_sales.drop("Fuel_Economy", axis = 1)
car_sales.drop("fuel_used_column", axis = 1)
car_sales.drop("Total_Fuel_Used", axis = 1)
```
